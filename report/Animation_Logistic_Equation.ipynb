```python
fig, ax = plt.subplots()

ax.plot([0, 1], [0, 1], 'black', lw=2)
line, = ax.plot([], [],color='r')
courbe, = ax.plot([], [], color='b')
x = np.linspace(0,1)

def init():
    line.set_data([], [])
    courbe.set_data([],[])
    return line,
    
def animate(i):
    line.set_data(zip(*lv.vectorization(0.05,0.05+i*0.05, 40)))
    courbe.set_data(x, lv.logistic(0.05+i*0.05,x))
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init,frames=82, interval=40)
```

```python
def logistic(r, x):
    return r*x*(1-x)
```

```python
n = 10000
r = np.linspace(2.5, 4.0, n)
iterations = 1000
last = 100
y = 1e-5 * np.ones(n)
```

```python
fig, ax = plt.subplots()

for i in range(iterations):
    y = logistic(r, y)
    if i >= (iterations - last):
        ax.plot(r, x, alpha=.25)
ax.set_xlim(2.5, 4)
ax.set_title("Bifurcation diagram")

plt.show()
```
